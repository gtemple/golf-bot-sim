```python
def calc_base(rank, exponent=0.4):
    top = 99.0
    bottom = 72.0
    TOP_N = 1000
    t = (rank - 1) / (TOP_N - 1)
    base = top - (top - bottom) * (t ** exponent)
    return base

print(f"Rank 1: {calc_base(1)}")
print(f"Rank 5: {calc_base(5)}")
print(f"Rank 10: {calc_base(10)}")
print(f"Rank 20: {calc_base(20)}")
print(f"Rank 50: {calc_base(50)}")
```
